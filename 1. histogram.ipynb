```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
sns.set(rc={'figure.figsize':(10,8)})
```

```python
df = pd.read_csv("pseudo_facebook.tsv", sep = '\t')
df.head()
```

## Histogram
[Seaborn Histogram](https://seaborn.pydata.org/generated/seaborn.distplot.html)

Flexibly plot a univariate distribution of observations.

This function combines the matplotlib hist function (with automatic calculation of a good default bin size) with the seaborn [kdeplot()](https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot) and [rugplot()](https://seaborn.pydata.org/generated/seaborn.rugplot.html#seaborn.rugplot) functions. It can also fit __scipy.stats__ distributions and plot the estimated PDF over the data.



```python
# default with kde
sns.distplot(df['dob_day']);

plt.title('Histogram of date of birth');
plt.xlabel('Date of birth');
plt.ylabel('Count');
```

```python
# histogram without kde
sns.distplot(df['dob_day'], kde = False);

plt.title('Histogram of date of birth');
plt.xlabel('Date of birth');
plt.ylabel('Count');
```

When drawing histograms, the main choice you have is the number of bins to use and where to place them. distplot() uses a simple rule to make a good guess for what the right number is by default, but trying more or fewer bins might reveal other features in the data:

### 1. Providing the number of bins

```python
# histogram without kde
sns.distplot(df['dob_day'], kde = False, bins = 10);

plt.title('Histogram of date of birth');
plt.xlabel('Date of birth');
plt.ylabel('Count');
```

## providing binwidth/bins/limit

```python
bins = np.arange(0, 120, 20)
bins
```

```python
sns.distplot(df['age'], kde = False, bins = bins);

plt.title('Histogram of date of birth');
plt.xlabel('Date of birth');
plt.ylabel('Count');
```

```python
# histogram without kde but with rugplot
sns.distplot(df['dob_day'], kde = False, rug = True);

plt.title('Histogram of date of birth');
plt.xlabel('Date of birth');
plt.ylabel('Count');
```

Plot the distribution with a kernel density estimate and rug plot but without histogram:

```python
sns.distplot(df['dob_day'], hist = False, kde = True, rug = True);

plt.title('kde and rugplot of date of birth');
plt.xlabel('Date of birth');
plt.ylabel('Count');
```

Plot the distribution with a histogram and maximum likelihood gaussian distribution fit.

```python
from scipy.stats import norm

sns.distplot(df['dob_day'], kde = False, fit = norm);

plt.title('Histogram of date of birth');
plt.xlabel('Date of birth');
plt.ylabel('Count');
```

Plot the distribution on the vertical axis

```python
# histogram without kde
sns.distplot(df['dob_day'], kde = False, vertical = True);

plt.title('Histogram of date of birth');
plt.xlabel('Date of birth');
plt.ylabel('Count');
```

Change the color of all the plot elements.

```python
sns.set_color_codes()
sns.distplot(df['dob_day'], color="y");
```

Pass specific parameters to the underlying plot functions:

```python
sns.distplot(df['dob_day'], rug = True, rug_kws = {"color":'g'},
            kde_kws = {'color':'k', "lw": 3, "label": "KDE"},
            hist_kws={"histtype": "step", "linewidth": 3,"alpha": 1, "color": "g"});

plt.title('Histogram of date of birth');
plt.xlabel('Date of birth');
plt.ylabel('Count');
```

# [Kernel Density Plot](https://datavizcatalogue.com/methods/density_plot.html)

![](kernel.PNG)



A __[kernel smoother](https://en.wikipedia.org/wiki/Kernel_smoother)__ is a statistical technique to estimate a real valued functionas the weighted average of neighboring observed data. The weight is defined by the kernel, such that closer points are given higher weights. The estimated function is smooth, and the level of smoothness is set by a single parameter.
